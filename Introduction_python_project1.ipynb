```python
missing_count = df['Y2014'].isnull().sum()

# Calculate the total number of entries in the column
total_entries = len(df['Y2014'])

# Calculate the percentage of missing data
percentage_missing = (missing_count / total_entries) * 100

# Round the percentage to 3 decimal places
percentage_missing = round(percentage_missing, 3)

print("Total number of missing data in column Y2014:", missing_count)
print("Percentage of missing data in column Y2014:", percentage_missing, "%")
```

```python
madagascar_protein_2015 = df[df['Area'] == 'Madagascar']['Y2015']

# Calculate the total protein supply quantity in Madagascar for 2015
total_protein_2015 = madagascar_protein_2015.sum()

print("Total protein supply quantity in Madagascar in 2015:", total_protein_2015)
```

```python
subset_df = df[['Y2017', 'Area']]

grouped = subset_df.groupby('Area')['Y2017'].sum().reset_index()

sorted_grouped = grouped.sort_values(by='Y2017')

seventh_lowest_area = sorted_grouped.iloc[6]['Area']

print(seventh_lowest_area)
```
